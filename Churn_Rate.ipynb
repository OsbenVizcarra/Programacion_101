```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd

df = pd.read_csv('/content/Telco-Customer-Churn.csv')
df.head()

# Data Exploration and Analysis

# 1. Churn Rate
print(f"Overall Churn Rate: {df['Churn'].value_counts(normalize=True) * 100}")

# 2. Impact of Contract Type
print("\nChurn Rate by Contract Type:")
print(df.groupby('Contract')['Churn'].value_counts(normalize=True))

# Visualizations
plt.figure(figsize=(8, 6))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Rate by Contract Type')
plt.show()

# 3. Impact of Tenure
plt.figure(figsize=(10, 6))
sns.histplot(x='tenure', hue='Churn', data=df, kde=True)
plt.title('Customer Tenure vs. Churn')
plt.show()


# 4. Impact of Services (e.g., Internet Service, Online Security, etc.)
# Example for Internet Service:
print("\nChurn Rate by Internet Service:")
print(df.groupby('InternetService')['Churn'].value_counts(normalize=True))

plt.figure(figsize=(8, 6))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title('Churn Rate by Internet Service')
plt.show()


# 5.  Identify potential customer segments

# Example:  Grouping by Contract and Internet Service
segment_analysis = df.groupby(['Contract', 'InternetService'])['Churn'].value_counts(normalize=True).unstack()
print("\nChurn Rate by Contract and Internet Service:")
segment_analysis

#  Potential Hypothesis Generation:

# * Hypothesis 1:  Customers with Month-to-Month contracts have a higher churn rate.
# * Hypothesis 2:  Customers with Fiber optic internet service show higher churn than DSL or no internet service.
# * Hypothesis 3:  Customers with shorter tenure (e.g., less than 6 months) are more likely to churn.
# * Hypothesis 4:  Customers with specific combinations of services (e.g. fiber optic and no online security) are prone to churn.

# Further actions:
# * Analyze the impact of other variables
# * Explore demographics (gender, partner, dependents)
# * Investigate payment methods
# * Consider the interactions between different variables


# Example Actions for Retention:
# * Offer better incentives or contracts for month-to-month customers
# * Improve customer experience with specific internet services
# * Implement loyalty programs
# * Targeted customer service for those showing signs of churn.
```
